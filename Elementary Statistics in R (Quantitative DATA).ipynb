## Cumulative Relative Frequency Distribution

```r
duration.cumrelfreq = duration.cumfreq / nrow(faithful)
duration.cumrelfreq
```

```r
old = options(digits=2)
duration.cumrelfreq
```

```r
old = options(digits=2) 
cbind(duration.cumfreq, duration.cumrelfreq)
```

```r
cumrelfreq0 = c(0, duration.cumrelfreq) 
plot(breaks, cumrelfreq0, 
   main="Old Faithful Eruptions",  # main title 
   xlab="Duration minutes", 
   ylab="Cumulative eruption proportion") 
lines(breaks, cumrelfreq0)        # join the points
```

**Alternative Solution**
We create an interpolate function **_*Fn*_** with the built-in function ecdf. Then we plot **_*Fn*_** right away. There is no need to compute the cumulative frequency distribution a **_*priori*_**.

```r
Fn = ecdf(duration)
plot(Fn, 
   main="Old Faithful Eruptions", 
   xlab="Duration minutes", 
   ylab="Cumulative eruption proportion")
```

## Stem-and-Leaf Plot
A **stem-and-leaf plot** of a quantitative variable is a textual graph that classifies data items according to their most significant numeric digits. In addition, we often merge each alternating row with its next row in order to simplify the graph for readability

We can generate a *linear regression model* of the two variables with the **_*lm*_** function, and then draw a trend line with **_*abline*_**.

```r
options(repr.plot.width=7, repr.plot.height=7)# CHANGE The size of graph
plot(duration, wait,            # plot the variables 
   xlab="Eruption duration",        # xaxis label 
   ylab="Time waited")              # yaxis label
abline(lm(wait ~ duration))
```
